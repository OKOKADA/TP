## Tranformations

```python
# Séparation des données
Y = df['Robot_ProtectiveStop'].values
X = df.drop(columns=['Timestamp', 'Num']).values
```

```python
# Division de la dataset en entrainement et en test

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42) # 80% training and 20% test
```

```python
# Equilibrage des classes du trainset uniquement
from imblearn.over_sampling import SMOTE
from collections import Counter
import joblib


from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42) # 80% training and 20% test

#  Vérification de la distribution avant équilibrage
print("Distribution des classes AVANT équilibrage:", Counter(Y_train))

#  Application de SMOTE pour générer des échantillons synthétiques
smote = SMOTE(random_state=42)
X_train_smote, Y_train_smote = smote.fit_resample(X_train, Y_train)

#  Vérification de la distribution après équilibrage
print("Distribution des classes APRÈS SMOTE :", Counter(Y_train_smote))

# Vérification des dimensions
print("Taille de X_train AVANT :", X_train.shape)
print("Taille de X_train APRÈS SMOTE :", X_train_smote.shape)
```

```python
# Sauvegarder l’objet SMOTE avec joblib
joblib.dump(smote, 'smote_model.pkl')
print("Objet SMOTE sauvegardé sous 'smote_model.pkl'")
```

```python
# Normalisation des données equilibrées(trainset equilibré)
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from joblib import dump, load



def normalisation(X, scaler_par):
  normalized_X = scaler_par.fit_transform(X)
  return normalized_X

scaler = MinMaxScaler()

X_train = normalisation(X_train_smote, scaler)
X_test = normalisation(X_test, scaler)
```

```python
# Sauvegarder l’objet normalisation avec joblib
joblib.dump(normalisation, 'normalisation_model.pkl')
print("Objet normalisation sauvegardé sous 'normalisation_model.pkl'")
```

```python
# Distribution des classes dans X_test
print("Distribution des classes dans X_test :", Counter(Y_test))
```

```python
# Sequencement des données d'entrainement et de test
def create_time_sequences(X, Y, seq_length=10):
    """
    Transforme les données en séquences temporelles.

    :param X: numpy array, les features
    :param Y: numpy array, les labels
    :param seq_length: int, la taille des séquences
    :return: (X_seq, Y_seq) les nouvelles données séquencées
    """
    seqX, seqY = [], []
    for i in range(len(Y) - seq_length):
        seqX.append(X[i:(i + seq_length)])  # Prend une séquence de longueur seq_length
        seqY.append(Y[i + seq_length])  # La sortie correspond au point suivant dans la séquence

    return np.array(seqX), np.array(seqY)

# Transformation des données d'entraînement et de test
seq_length = 10
X_train_seq, Y_train_seq = create_time_sequences(X_train, Y_train, seq_length)
X_test_seq, Y_test_seq = create_time_sequences(X_test, Y_test, seq_length)

# Vérification des nouvelles dimensions
print("Shape of X_train_seq:", X_train_seq.shape)
print("Shape of Y_train_seq:", Y_train_seq.shape)
print("Shape of X_test_seq:", X_test_seq.shape)
print("Shape of Y_test_seq:", Y_test_seq.shape)
```

```python
# Sauvegarder l’objet normalisation avec joblib
joblib.dump(create_time_sequences, 'create_time_sequences_model.pkl')
print("Objet create_time_sequences sauvegardé sous 'create_time_sequences_model.pkl'")
```

# Entrainement des modèles



```python
# Entrainement du modèle LSTM

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

# Supposons que X et y sont tes données prétraitées (séquences et cibles)
# X shape : (n_samples, seq_length=10, nombre_feature=22)
# y shape : (n_samples,)

# Séparer en train et test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Définir le modèle LSTM
seq_length = 10
nombre_feature = 22

model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(seq_length, nombre_feature)))
model.add(Dropout(0.2))  # Ajouté pour éviter le surapprentissage
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1, activation='sigmoid'))  # Binaire (ajuste si multiclasse)

# Compiler le modèle
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])  # Binaire ici

# Entraîner le modèle
model.fit(X_train_seq, Y_train_seq, epochs=20, batch_size=32, validation_data=(X_test_seq, Y_test_seq), verbose=1)

# Évaluer le modèle
loss_lstm, accuracy_lstm = model.evaluate(X_test_seq, Y_test_seq)
print(f"LSTM Test Loss: {loss_lstm}")
print(f"LSTM Test Accuracy: {accuracy_lstm}")

# Prédire pour obtenir le rapport de classification
y_pred_lstm = (model.predict(X_test_seq) > 0.5).astype(int)  # Seuil à 0.5 pour binaire
report_lstm = classification_report(Y_test_seq, y_pred_lstm)
print("LSTM Classification Report:")
print(report_lstm)
```

```python
# Tester différentes combinaisons d’hyperparamètres pour le modèle LSTM
hyperparams = [
    {'units': 50, 'dropout_rate': 0.2, 'optimizer': 'adam'},
    {'units': 100, 'dropout_rate': 0.3, 'optimizer': 'rmsprop'},
    {'units': 75, 'dropout_rate': 0.2, 'optimizer': 'sgd'}
]


# Compiler le modèle
Losses, Accuracies = [], []

for param in hyperparams:
    print(f"\nTest avec {param}")
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(seq_length, nombre_feature)))
    model.add(Dropout(param['dropout_rate']))  # Ajouté pour éviter le surapprentissage
    model.add(LSTM(units=param['units']))
    model.add(Dropout(param['dropout_rate']))
    model.add(Dense(units=1, activation='sigmoid'))
    model.compile(optimizer=param['optimizer'], loss='binary_crossentropy', metrics=['accuracy'])


    model.fit(X_train_seq,
              Y_train_seq, epochs=20,
              batch_size=32, validation_data=(X_test_seq, Y_test_seq),
              verbose=1)
    # Évaluation
    loss, accuracy = model.evaluate(X_test_seq, Y_test_seq)
    Losses.append(loss)
    Accuracies.append(accuracy)
    print(f"Loss: {loss}, Accuracy: {accuracy}")
    # Prédictions pour métriques supplémentaires
    y_pred = (model.predict(X_test_seq) > 0.5).astype(int)  # Si binaire
    print(classification_report(Y_test_seq, y_pred))
```
